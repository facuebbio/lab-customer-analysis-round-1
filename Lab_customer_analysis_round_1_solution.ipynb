```python
# Understand the meaning of all columns:

# customer: means that every client or customer is identified with a serial number composed by letters and numbers.
# state: means the province or state where the customer lives.
# gender: means if the customer is Male, Female or NaN
# education: means the maximun education a customer has completed until that date.
# customer lifetime value: means estimated value that an the insurance company expects to derive from each customer over the entire duration of their relationship.
# income: means the amount of money each customer perceive per year.
# monthly premium auto: means the monthly payment made by a customer to maintain an auto insurance policy.
# complaints: means the count or total number of claims/complaints that have been submitted by each customer.
# policy type: means the contract between an the insurance company and the customers.
# vehicle class: means what type of vehicle (e.g. Four Door Car) are each of the customers covering with the insurance they're paying for.
# total claim amount: means the sum of money that the insurance company is expected to pay out to cover the costs associated with a covered loss or event.
```

```python
# Perform the data cleaning operations mentioned so far in class
```

```python
# Delete the column education and the number of open complaints from the dataframe.

final_dataset = dataset.drop(columns=["education","number of open complaints"])
final_dataset
```

```python
def function_percentage(x):
    if "%" in str(x):
        x = x.replace("%", "")
        return float(x) * 100
    else:
        return float(x) * 100 if pd.notna(x) else 0

final_dataset["customer lifetime value"] = final_dataset["customer lifetime value"].apply(function_percentage)
final_dataset
```

```python
final_dataset["customer lifetime value"] = np.float32(final_dataset["customer lifetime value"])*100
final_dataset
```
